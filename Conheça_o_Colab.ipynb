## Ciência de dados
</div>

Com o Colab, você pode aproveitar todo o potencial das conhecidas bibliotecas Python para analisar e ver dados. A célula de códigos abaixo usa <strong>numpy</strong> para gerar dados aleatórios e <strong>matplotlib</strong> para visualizá-los. Para editar o código, basta clicar na célula e começar a editar.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
from google.colab import files
uploaded = files.upload()
```

```
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```
df.info()
```

```

df.describe()
```

```

df.isnull().sum()
```

```
df["type"].value_counts()
```

```
df["type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Distribuição de Filmes e Séries")
```

```
df["country"].value_counts().head(10)
```

```
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 países com mais conteúdo")
plt.xticks(rotation=45)
```

```
df["release_year"].value_counts().sort_index()
```

```
df["release_year"].value_counts().sort_index().plot()
plt.title("Conteúdo ao longo dos anos")
plt.xlabel("Ano")
plt.ylabel("Quantidade")
```

```
df.groupby(["release_year", "type"]).size().unstack()
```

```
df.groupby(["release_year", "type"]).size().unstack().plot()
plt.title("Filmes vs Séries ao longo dos anos")
plt.xlabel("Ano")
plt.ylabel("Quantidade")
```
